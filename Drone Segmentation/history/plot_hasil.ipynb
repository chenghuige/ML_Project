```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
unet_b3_1 = pd.read_csv('FPN-efficenet-b3_cross_entropy.csv')
unet_b3_1.drop('Unnamed: 0', axis=1, inplace=True)
unet_b3_1 = unet_b3_1.dropna()
unet_b3_1
```

```python
unet_b3_2 = pd.read_csv('FPN-efficenet-b3_lovasz.csv')
unet_b3_2.drop('Unnamed: 0', axis=1, inplace=True)
unet_b3_2 = unet_b3_2.dropna()
unet_b3_2
```

```python
unet_b3 = pd.concat([unet_b3_1, unet_b3_2], ignore_index=True)
unet_b3
```

```python
plt.plot(np.arange(1, len(unet_b3)+1), unet_b3['val_miou'], marker = 'o', label='Val mIoU')
plt.plot(np.arange(1, len(unet_b3)+1), unet_b3['train_miou'], marker= '^', label='Train mIoU')
plt.xlim(0, len(unet_b3)+3); plt.ylim(0, max(unet_b3['val_miou'])+0.05)
plt.vlines(len(unet_b3_1), 0.03,  max(unet_b3['val_miou'])+0.02)
plt.text(25, 0.35, 'Training Phase 2')
plt.text(8, 0.12, 'Training Phase 1')
plt.title('FPN EfficientNet-B3')
plt.grid(axis='y')
plt.legend()
plt.show()
```

```python
plt.plot(np.arange(1, len(unet_b3)+1), unet_b3['val_loss'], marker = 'o', label='Val Loss')
plt.plot(np.arange(1, len(unet_b3)+1), unet_b3['train_loss'], marker= '^', label='Train Loss')
plt.xlim(0, len(unet_b3)+3); plt.ylim(0, max(unet_b3['val_loss'])+0.05)
plt.vlines(len(unet_b3_1), 0.03,  max(unet_b3['val_loss'])+0.02)
plt.text(25, 0.63, 'Training Phase 2')
plt.text(7, 0.85, 'Training Phase 1')
plt.title('FPN EfficientNet-B3')
plt.grid(axis='y')
plt.legend()
plt.show()
```

```python
plt.plot(np.arange(1, len(unet_b3)+1), unet_b3['val_acc'], marker = 'o', label='Val Acc')
plt.plot(np.arange(1, len(unet_b3)+1), unet_b3['train_acc'], marker= '^', label='Train Acc')
plt.xlim(0, len(unet_b3)+3); plt.ylim(0, max(unet_b3['val_acc'])+0.07)
plt.vlines(len(unet_b3_1), 0.03,  max(unet_b3['val_acc'])+0.05)
plt.text(25, 0.61, 'Training Phase 2')
plt.text(7, 0.61, 'Training Phase 1')
plt.title('FPN EfficientNet-B3')
plt.grid(axis='y')
plt.legend()
plt.show()
```
